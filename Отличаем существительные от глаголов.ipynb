```python
import numpy as np
```

```python
# Обертка чтобы использовать обычные мат. операнды над numpy матрицами


class Matrix:
    def __init__(self, python_array=None, ndarray=None):
        if python_array:
            self.array = np.array(python_array)
        else:
            self.array = ndarray
    
    def __mul__(self, other):
        return Matrix(ndarray=np.matmul(self.array, other.array))

    def __pow__(self, n):
        return Matrix(ndarray=np.linalg.matrix_power(self.array, n))

    def t(self):
        return Matrix(ndarray=np.transpose(self.array))

    def __str__(self):
        return str(self.array)

    def __repr__(self):
        return repr(self.array)

    def __add__(self, other):
        return Matrix(ndarray=self.array + other.array)

    def as_number(self):
        assert all(map(lambda x: x == 1, self.array.shape)), f'Matrix with shape={self.array.shape} нельзя перевести в число'
        return self.array[0]  # TODO: написать метод нормально
        
    def __iter__(self):
        return iter(self.array)
```

```python
def count_regression(X, Y):
    """ Y = XB + E, здесь B - искомые коофициенты регрессии, Е - вектор ошибок регрессии """
    return (X.t() * X) ** -1 * X.t() * Y
```
